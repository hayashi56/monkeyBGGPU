```python
import numpy as np
import matplotlib.pyplot as plt

synapsepotentialdata_file_MSND1 = 'MSN_D1firingrate.dat'
synapsepotentialdata_file_MSND2 = 'MSN_D2firingrate.dat'
data_file_MSND1 = 'MSN_D1.dat'
data_file_MSND2 = 'MSN_D2.dat'
synapsepotentialdata_file_FSI = 'FSIfiringrate.dat'
synapsepotentialdata_file_STN = 'STNfiringrate.dat'
synapsepotentialdata_file_GPe = 'GPefiringrate.dat'
synapsepotentialdata_file_PTN = 'PTNfiringrate.dat'
alpha_data_file_MSND1 = 'MSN_D1_debag.dat'
alpha_data_file_MSND2 = 'MSN_D2_debag.dat'
synapsepotentialdata_file_FSI = 'FSIfiringrate.dat'
synapsepotentialdata_file_STN = 'STNfiringrate.dat'
synapsepotentialdata_file_GPe = 'GPefiringrate.dat'
synapsepotentialdata_file_GPi = 'GPifiringrate.dat'
synapsepotentialdata_file_PTN = 'PTNfiringrate.dat'
synapsepotentialdata_file_PTI = 'PTIfiringrate.dat'
synapsepotentialdata_file_Th = 'Thfiringrate.dat'
synapsepotentialdata_file_CMPf = 'CMPffiringrate.dat'
synapsepotentialdata_file_PSA = 'PSNfiringrate.dat'
MSND1potential = np.loadtxt(data_file_MSND1)
MSND2potential = np.loadtxt(data_file_MSND2)
MSND1synapsepotential = np.loadtxt(synapsepotentialdata_file_MSND1)
MSND2synapsepotential = np.loadtxt(synapsepotentialdata_file_MSND2)
FSIsynapsepotential = np.loadtxt(synapsepotentialdata_file_FSI)
STNsynapsepotential = np.loadtxt(synapsepotentialdata_file_STN)
GPesynapsepotential = np.loadtxt(synapsepotentialdata_file_GPe)
GPisynapsepotential = np.loadtxt(synapsepotentialdata_file_GPi)
PTNsynapsepotential = np.loadtxt(synapsepotentialdata_file_PTN)
PTIsynapsepotential = np.loadtxt(synapsepotentialdata_file_PTI)
Thsynapsepotential = np.loadtxt(synapsepotentialdata_file_Th)
CMPfsynapsepotential = np.loadtxt(synapsepotentialdata_file_CMPf)
PSAsynapsepotential = np.loadtxt(synapsepotentialdata_file_PSA)
```
