```python
import numpy as np
import matplotlib.pyplot as plt

frequencydata_file1_GPi = 'GPifiringrate_chanel1.dat'    ## GPiの発火率についてのデータをファイルから読み込み
frequencydata_file2_GPi = 'GPifiringrate_chanel2.dat'    ## GPiの発火率についてのデータをファイルから読み込み

GPifrequency1 = np.loadtxt(frequencydata_file1_GPi)     ## GPiの発火率についての読み込んだデータをnumpyのに変換
GPifrequency2 = np.loadtxt(frequencydata_file2_GPi)     ## GPiの発火率についての読み込んだデータをnumpyのに変換
```

```python
e1 = 1 - GPifrequency1 / 44.663635    ##chanel1(選択肢1)の発火率の安静時の発火率との変化度合いを計算
e2 = 1 - GPifrequency2 / 44.663635    ##chanel2(選択肢2)の発火率の安静時の発火率との変化度合いを計算
e_1, e_2 = e1.copy(), e2.copy()
e = e_2 - e_1    ##eの差を計算
f = -1 * ( - 0.01 > e ) + 1 * ( 0.01 < e ) + 0 * ( - 0.01 <= e ) * ( 0.01 >= e )
```

```python
map_firingrate = f.reshape( [ 20, 20 ] )    ##mapの形に変える
tick = range ( 0, 20, 1 )    ##軸の分割

fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(2, 2, 1)
ax.set_xlabel("chanel1 firing rate")
ax.set_ylabel("chanel2 firing rate")
ax.set_xticks ( tick )
ax.set_yticks ( tick )
map = ax.pcolor ( map_firingrate )
fig.colorbar ( map, ax = ax )
plt.show()
```
