```python
import numpy as np
import matplotlib.pyplot as plt

spikedata_file1_MSN_D1 = 'MSN_D1spike.dat'
spikedata_file1_MSN_D2 = 'MSN_D2spike.dat'
spikedata_file1_FSI    = 'FSIspike.dat'
spikedata_file1_STN    = 'STNspike.dat'
spikedata_file1_GPe    = 'GPespike.dat'
spikedata_file1_GPi    = 'GPispike.dat'
spikedata_file1_SNc    = 'SNcspike.dat'
spikedata_file1_PTI    = 'PTIspike.dat'
spikedata_file1_PTN    = 'PTNspike.dat'
spikedata_file1_PSA    = 'PSAspike.dat'
spikedata_file1_Th     = 'Thspike.dat'
spikedata_file1_CMPf   = 'CMPfspike.dat'

MSN_D1spike1 = np.loadtxt(spikedata_file1_MSN_D1)
MSN_D2spike1 = np.loadtxt(spikedata_file1_MSN_D2)
FSIspike1 = np.loadtxt(spikedata_file1_FSI)
STNspike1 = np.loadtxt(spikedata_file1_STN)
GPespike1 = np.loadtxt(spikedata_file1_GPe)
GPispike1 = np.loadtxt(spikedata_file1_GPi)
SNcspike1 = np.loadtxt(spikedata_file1_SNc)
PTIspike1 = np.loadtxt(spikedata_file1_PTI)
PTNspike1 = np.loadtxt(spikedata_file1_PTN)
PSAspike1 = np.loadtxt(spikedata_file1_PSA)
Thspike1 = np.loadtxt(spikedata_file1_Th)
CMPfspike1 = np.loadtxt(spikedata_file1_CMPf)

frequencydata_file1_MSN_D1 = 'MSN_D1firingrate.dat'
frequencydata_file1_MSN_D2 = 'MSN_D2firingrate.dat'
frequencydata_file1_FSI = 'FSIfiringrate.dat'
frequencydata_file1_STN = 'STNfiringrate.dat'
frequencydata_file1_GPe = 'GPefiringrate.dat'
frequencydata_file1_GPi = 'GPifiringrate.dat'
frequencydata_file1_SNc = 'SNcfiringrate.dat'
frequencydata_file1_PTI = 'PTIfiringrate.dat'
frequencydata_file1_PTN = 'PTNfiringrate.dat'
frequencydata_file1_PSA = 'PSAfiringrate.dat'
frequencydata_file1_Th = 'Thfiringrate.dat'
frequencydata_file1_CMPf = 'CMPffiringrate.dat'

MSN_D1frequency1 = np.loadtxt(frequencydata_file1_MSN_D1)
MSN_D2frequency1 = np.loadtxt(frequencydata_file1_MSN_D2)
FSIfrequency1 = np.loadtxt(frequencydata_file1_FSI)
STNfrequency1 = np.loadtxt(frequencydata_file1_STN)
GPefrequency1 = np.loadtxt(frequencydata_file1_GPe)
GPifrequency1 = np.loadtxt(frequencydata_file1_GPi)
SNcfrequency1 = np.loadtxt(frequencydata_file1_SNc)
PTIfrequency1 = np.loadtxt(frequencydata_file1_PTI)
PTNfrequency1 = np.loadtxt(frequencydata_file1_PTN)
PSAfrequency1 = np.loadtxt(frequencydata_file1_PSA)
Thfrequency1 = np.loadtxt(frequencydata_file1_Th)
CMPffrequency1 = np.loadtxt(frequencydata_file1_CMPf)

print("MSND1: " + str(MSN_D1frequency1))
print("MSND2: " + str(MSN_D2frequency1))
print("FSI:   " + str(FSIfrequency1))
print("STN:   " + str(STNfrequency1))
print("GPe:   " + str(GPefrequency1))
print("GPi:   " + str(GPifrequency1))
print("SNc:   " + str(SNcfrequency1))
print("PTI:   " + str(PTIfrequency1))
print("PTN:   " + str(PTNfrequency1))
print("PSA:   " + str(PSAfrequency1))
print("Th:    " + str(Thfrequency1))
print("CMPf:  " + str(CMPffrequency1))
```
